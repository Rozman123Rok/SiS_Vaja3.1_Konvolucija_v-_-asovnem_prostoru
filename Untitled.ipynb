```python
# Rok Rozman Konvolucija v ƒçasovnem prostoru

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
import sounddevice as sd
from scipy.io import wavfile
```

```python
def konv_cas_mono(signal, impulz):
    x = signal
    h = impulz
    N = x.shape[0]+h.shape[0]-1
    y = np.zeros(N)
    for n in range(N):
        #print('....')
        #print(f'n={n}')
    
        for k in range(max(n-h.shape[0]+1,0), min(n+1, x.shape[0])):
            #print(f'    k={k}')
            #print(f'        n-k={n-k}')
            y[n] = y[n]+x[k]*h[n-k]
            
    y2 = np.convolve(x, h)
    plt.figure()

    plt.plot(y, linewidth=2, label='for zanka')
    #plt.plot(y2, 'r:', linewidth=2, label='conv')
    plt.xlabel('vzorci')
    plt.ylabel('amplituda')
    plt.tight_layout()
    plt.legend(loc='best')

    plt.show()
    
    plt.figure()

    #plt.plot(y, linewidth=2, label='for zanka')
    plt.plot(y2, 'r:', linewidth=2, label='conv')
    plt.xlabel('vzorci')
    plt.ylabel('amplituda')
    plt.tight_layout()
    plt.legend(loc='best')

    plt.show()
    
    y = y / np.max(np.abs(y)) # normalizirajmo
    y2 = y2 / np.max(np.abs(y2)) # normalizirajmo
        
    plt.figure()

    plt.plot(y, linewidth=2, label='for zanka')
    #plt.plot(y2, 'r:', linewidth=2, label='conv')
    plt.xlabel('vzorci')
    plt.ylabel('amplituda')
    plt.tight_layout()
    plt.legend(loc='best')

    plt.show()
    
    plt.figure()

    #plt.plot(y, linewidth=2, label='for zanka')
    plt.plot(y2, 'r:', linewidth=2, label='conv')
    plt.xlabel('vzorci')
    plt.ylabel('amplituda')
    plt.tight_layout()
    plt.legend(loc='best')

    plt.show()
    
```
